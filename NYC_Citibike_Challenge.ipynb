```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
data_path = "Resources/201908-citibike-tripdata.csv"
citibike_df = pd.read_csv(data_path)
citibike_df.head()
```

```python
# 2. Check the datatypes of your columns. 
citibike_df.dtypes
```

```python
# 3. Duplicate the'tripduration' column and convert the new 'tripduration' column to a datetime datatype.
citibike_df['tripduration_minutes'] = citibike_df['tripduration']
citibike_df['tripduration_minutes'] = pd.to_datetime(citibike_df['tripduration_minutes'], unit='s')
citibike_df.head()
```

```python
# 4. Check the datatypes of your columns. 
citibike_df.dtypes
```

```python
# Get list of columns columns
columns = citibike_df.columns.tolist()
columns
```

```python
# Reorder columns
citibike_df = citibike_df[['tripduration_minutes',
     'tripduration',
     'starttime',
     'stoptime',
     'start station id',
     'start station name',
     'start station latitude',
     'start station longitude',
     'end station id',
     'end station name',
     'end station latitude',
     'end station longitude',
     'bikeid',
     'usertype',
     'birth year',
     'gender']]
citibike_df.head()
```

```python
# 5. Export the DataFrame as a new CSV file without the index column
citibike_csv = citibike_df.to_csv('resources/201908-citibike-tripdata-converted.csv',mode = 'w', index=False)
```
